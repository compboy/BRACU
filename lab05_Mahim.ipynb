```python
#Task03; Load the dataset as dataframe using pandas

import pandas as pd
import numpy as np
import sklearn
glass = pd.read_csv('/content/sample_data/glass source classification dataset.csv')
glass.head(20)
```

```python
glass.isnull().sum()
```

```python
#Task04;Handle missing values if needed

from sklearn.impute import SimpleImputer

impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(glass[['Ca']])
glass['Ca']=impute.transform(glass[['Ca']])
print(glass.isnull().sum())
```

```python
glass.info()
```

```python
#Task05; Encode categorical features if needed

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
glass['Ba']=encoder.fit_transform(glass['Ba'])
print(glass[['Ba']].head(100))
glass['Fe']=encoder.fit_transform(glass['Fe'])
print(glass[['Fe']].head())
glass['Type']=encoder.fit_transform(glass['Type'])
print(glass[['Type']].head())
```

```python
#Task06; Scale all the values between 0-1 with proper scaling technique

from sklearn.model_selection import train_test_split

X,Y=glass.iloc[:,2:13],glass.iloc[:,13:]
X_train, X_test,Y_train, Y_test=train_test_split(X,Y,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
#Task07; Split the dataset into features and labels. Use your intuition to determine which column indicates the labels.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
print("Per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("Per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_scaled.max(axis=0)))
```
