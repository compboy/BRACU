```python
#Task03; Load the dataset as dataframe using pandas

import pandas as pd
import numpy as np
import sklearn
mashroom = pd.read_csv('/content/sample_data/mushroom edibility classification dataset.csv')
mashroom.head()
```

```python
mashroom.shape
```

```python
mashroom.isnull().sum()
```

```python
#Task04;Handle missing values if needed

from sklearn.impute import SimpleImputer

impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(mashroom[['cap-shape']], mashroom[['cap-color']])
mashroom['cap-shape']=impute.transform(mashroom[['cap-shape']])
mashroom['cap-color']=impute.transform(mashroom[['cap-color']])
print(mashroom.isnull().sum())
```

```python
#Task05; Encode categorical features if needed

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
mashroom['class']=encoder.fit_transform(mashroom['class'])
print(mashroom[['class']].head())
mashroom['bruises']=encoder.fit_transform(mashroom['bruises'])
print(mashroom[['bruises']].head())
```

```python
#Task06; Scale all the values between 0-1 with proper scaling technique

from sklearn.model_selection import train_test_split

X,Y=mashroom.iloc[:,2:13],mashroom.iloc[:,13:]
X_train, X_test,Y_train, Y_test=train_test_split(X,Y,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
#Task07; Split the dataset into features and labels. Use your intuition to determine which column indicates the labels.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
print("Per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("Per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_scaled.max(axis=0)))
```
